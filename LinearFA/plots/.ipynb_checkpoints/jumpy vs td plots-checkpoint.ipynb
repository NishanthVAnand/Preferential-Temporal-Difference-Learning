```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from matplotlib import colors
import matplotlib.backends.backend_pdf
import seaborn as sns
import pandas as pd
sns.set_style("whitegrid")
sns.set_palette("bright", color_codes=True)
```

```python
#pdf = matplotlib.backends.backend_pdf.PdfPages("jumpy_td_vs_td"+"_lr_"+str(lr)+"_env_"+str(env_len)+"_epi_"+str(episodes)+".pdf")
types = ["jumpy_td", "td"]
seed = [i for i in range(20)]
lr = 0.01
lamb = [0.0, 0.25, 0.5, 0.75, 0.9]
env_len = 5
episodes = 1000

fig, ax = plt.subplots(1,2, figsize=(12, 6))
fig.tight_layout()

row_plt = 0
col_plt = 0
for t in types:
    for l in lamb:       
        no_labels = []
        df_no_beta = pd.DataFrame(columns=["Episodes", "MSVE"])
        df_col1 = []
        df_col2 = []
        for se in seed:
            try:
                with open("../results/"+t+"_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se)+".pkl", "rb") as f:
                    tmpp = list(pickle.load(f))
                    df_col2.extend(tmpp)
                    df_col1.extend([i for i in range(50,1001,50)])
            except:
                print(t+"_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se))
        df_no_beta["Episodes"] = df_col1
        df_no_beta["MSVE"] = df_col2

        sns.lineplot(x="Episodes", y ="MSVE", data=df_no_beta, ax=ax[col_plt], ci=60, err_style="band")

    ax[col_plt].set_ylabel('')    
    ax[col_plt].set_xlabel('')
    ax[col_plt].set_ylim(0,0.8)

    ax[col_plt].set_title(t)

    col_plt += 1
    if col_plt == 2:
        col_plt = 0
        row_plt += 1

    fig.legend(lamb, loc='upper right')
    #pdf.savefig(fig)
    #pdf.close()
    
```

```python
#pdf = matplotlib.backends.backend_pdf.PdfPages("jumpy_td_vs_td"+"_lr_"+str(lr)+"_env_"+str(env_len)+"_epi_"+str(episodes)+".pdf")
fig, ax = plt.subplots(1,1, figsize=(8, 6))
#fig.delaxes(ax[1,3])
best_pair = [("jumpy_td", 0.5), ("td", 0.0)]
fig.tight_layout()
row_plt = 0
col_plt = 0
for t, l in best_pair:
    df_no_beta = pd.DataFrame(columns=["Episodes", "MSVE"])
    no_beta_rew_temp = []
    for se in seed:
        try:
            with open("../results/"+t+"_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se)+".pkl", "rb") as f:
                tmpp = list(pickle.load(f))
                no_beta_rew_temp.append(np.array(tmpp))
                df_col2.extend(tmpp)
                df_col1.extend([i for i in range(50,1001,50)])
        except:
            print(t+"_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se))
    df_no_beta["Episodes"] = df_col1
    df_no_beta["MSVE"] = df_col2
    no_labels.append(t+" lamb "+str(l))

    sns.lineplot(x="Episodes", y ="MSVE", data=df_no_beta, ci=60, err_style="band")
plt.legend(no_labels)
plt.show()
#pdf.savefig(fig)
#pdf.close()
```

```python
# find the P matrix
P_pi = np.zeros((11, 11))
for r_idx in range(P_pi.shape[0]):
    if r_idx < 5:
        P_pi[r_idx][r_idx+1] = 0.5
        P_pi[r_idx][r_idx+1+5] = 0.5
    elif r_idx > 5 and r_idx < 10:
        P_pi[r_idx][r_idx+1] = 0.5
        P_pi[r_idx][r_idx+1-5] = 0.5
    else:
        pass
r_exp = 1

# find the reward vector
r_pi = np.array([r_exp for i in range(5)] + [0] + [-r_exp for i in range(5-1)] + [0])

# find the true value
true_value = np.linalg.inv(np.eye(11) - 0.99 * P_pi).dot(r_pi)

#bias-variance

#pdf = matplotlib.backends.backend_pdf.PdfPages("jumpy_td_vs_td"+"_lr_"+str(lr)+"_env_"+str(env_len)+"_epi_"+str(episodes)+".pdf")
types = ["bias", "var"]
seed = [i for i in range(20)]
lr = 0.01
lamb = [0.0, 0.25, 0.5, 0.75, 0.9]
env_len = 5
episodes = 1000

df_bias = pd.DataFrame(columns=["Lambda"]+[str(i) for i in range(50,1001,50)])
df_var = pd.DataFrame(columns=["Lambda"]+[str(i) for i in range(50,1001,50)])

df_bias["Lambda"] = [l for l in lamb]
df_var["Lambda"] = [l for l in lamb]

for l in lamb:
    total_ = []
    for se in seed:
        try:
            with open("../results/jumpy_td_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se)+"_value_.pkl", "rb") as f:
                tmpp = pickle.load(f)
                total_.append(tmpp)
        except:
            print("../results/jumpy_td_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se)+"_value_.pkl")
    total_np = np.array(total_)
    curr_var = np.var(total_, axis=0)
    mean_var = list(np.mean(curr_var, axis=1))
    
    curr_mean = np.mean(total_np, axis=0)
    bias = list(np.mean((curr_mean - true_value)**2, axis=1)**0.5)

    df_var.loc[df_var["Lambda"]==l, 1:] = mean_var
    df_bias.loc[df_bias["Lambda"]==l, 1:] = bias

df_bias = df_bias.astype(float)
df_var = df_var.astype(float)
#pdf.savefig(fig)
#pdf.close()
```

```python
P_pi = np.zeros((11, 11))
for r_idx in range(P_pi.shape[0]):
    if r_idx < 5:
        P_pi[r_idx][r_idx+1] = 0.5
        P_pi[r_idx][r_idx+1+5] = 0.5
    elif r_idx > 5 and r_idx < 10:
        P_pi[r_idx][r_idx+1] = 0.5
        P_pi[r_idx][r_idx+1-5] = 0.5
    else:
        pass
r_exp = 1
r_pi = np.array([r_exp for i in range(5)] + [0] + [-r_exp for i in range(5-1)] + [0])
true_value = np.linalg.inv(np.eye(11) - 0.99 * P_pi).dot(r_pi)

#bias-variance

#pdf = matplotlib.backends.backend_pdf.PdfPages("jumpy_td_vs_td"+"_lr_"+str(lr)+"_env_"+str(env_len)+"_epi_"+str(episodes)+".pdf")
types = ["bias", "var"]
seed = [i for i in range(20)]
lr = 0.01
lamb = [0.0, 0.25, 0.5, 0.75, 0.9]
env_len = 5
episodes = 1000

df_bias_org = pd.DataFrame(columns=["Lambda"]+[str(i) for i in range(50,1001,50)])
df_var_org = pd.DataFrame(columns=["Lambda"]+[str(i) for i in range(50,1001,50)])

df_bias_org["Lambda"] = [l for l in lamb]
df_var_org["Lambda"] = [l for l in lamb]

for l in lamb:
    org_total_ = []
    for se in seed:
        try:
            with open("../results/td_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se)+"_value_.pkl", "rb") as f:
                tmpp = pickle.load(f)
                org_total_.append(tmpp)
        except:
            print("../results/td_lamb_"+str(l)+"_lr_"+str(lr)+"_episodes_"+str(episodes)+"_env_len_"+str(env_len)+"_seed_"+str(se)+"_value_.pkl")
    org_total_np = np.array(org_total_)
    org_curr_var = np.var(org_total_, axis=0)
    org_mean_var = list(np.mean(org_curr_var, axis=1))
    
    org_curr_mean = np.mean(org_total_np, axis=0)
    org_bias = list(np.mean((org_curr_mean - true_value)**2, axis=1)**0.5)

    df_var_org.loc[df_var["Lambda"]==l, 1:] = org_mean_var
    df_bias_org.loc[df_bias["Lambda"]==l, 1:] = org_bias

df_bias_org = df_bias_org.astype(float)
df_var_org = df_var_org.astype(float)
#pdf.savefig(fig)
#pdf.close()
```

```python
fig, ax = plt.subplots(4,5, figsize=(20, 12), sharex='col', sharey='row')
fig.tight_layout(h_pad=1)

row_plt = 0
col_plt = 0

for i in range(1, 21):    
    sns.lineplot(x="Lambda", y =str(i*50), data=df_bias, ax=ax[row_plt, col_plt], color='c')
    sns.lineplot(x="Lambda", y =str(i*50), data=df_var, ax=ax[row_plt, col_plt], color='b')
    
    sns.lineplot(x="Lambda", y =str(i*50), data=df_bias_org, ax=ax[row_plt, col_plt], color='k')
    sns.lineplot(x="Lambda", y =str(i*50), data=df_var_org, ax=ax[row_plt, col_plt], color='r')
    
    ax[row_plt, col_plt].set_ylabel('')    
    ax[row_plt, col_plt].set_xlabel('')
    ax[row_plt, col_plt].set_ylim(0,0.8)
    ax[row_plt, col_plt].set_title("Episode: "+str(i*50), fontdict={'fontsize':13})
    ax[row_plt, col_plt].set_yticks(np.arange(0, 1, step=0.2))
    ax[row_plt, col_plt].tick_params(labelsize=13)

    col_plt+=1
    if col_plt >= 5:
        row_plt+=1
        col_plt=0

fig.legend(["Bias - additive lambda", "Variance - additive lambda", "Bias - original lambda", "Variance - original lambda"], fontsize='large')
plt.show()
```

```python
def getVec(value):
    pos_x = []
    pos_y = []
    dir_x = []
    dir_y = []

    for i, val in enumerate(value):
        if i < 5:
            u_next = value[i+1]
            l_next = value[i+5+1]
            if u_next >= l_next:
                pos_x.append(i+0.5)
                pos_y.append(1.5)
                dir_x.append(1)
                dir_y.append(0)
            if u_next <= l_next:
                pos_x.append(i+0.5)
                pos_y.append(1.5)
                dir_x.append(0.707)
                dir_y.append(-0.707)
        if i > 5 and i < 10:
            u_next = value[i+1-5]
            l_next = value[i+1]
            if u_next >= l_next:
                pos_x.append(i%5+0.5)
                pos_y.append(0.5)
                dir_x.append(0.707)
                dir_y.append(0.707)
            if u_next <= l_next:
                pos_x.append(i%5+0.5)
                pos_y.append(0.5)
                dir_x.append(1)
                dir_y.append(0)
    
    pos_x.extend([5.5, 5.5])
    pos_y.extend([0.5, 1.5])
    dir_x.extend([1,1])
    dir_y.extend([0,0])
                
    return np.array(pos_x), np.array(pos_y), np.array(dir_x), np.array(dir_y)
```

```python
lr = 0.01
best_pair = [("jumpy_td", 0.5), ("td", 0.0)]
seed = [i for i in range(20)]
chain_len = 5
num_chains = 2
col_data =  np.ones((2, chain_len+2))
col_data[0][0] = col_data[0][-1] = col_data[1][-1] = 0.6
fig, ax = plt.subplots(10,2, figsize=(10, 15))

cols = ["jumpy_td", "td"]
for axe, col in zip(ax[0], cols):
    axe.set_title(col, {'fontsize': 16})

for col, (t, l) in enumerate(best_pair):
    all_data = []
    for s in seed:
        with open("../results/"+t+"_lamb_"+str(l)+"_lr_0.01_episodes_10_env_len_5_seed_"+str(s)+"_value_.pkl", "rb") as f:
            data = pickle.load(f)
            all_data.append(data)
    for i in range(10):
        ax[i, col].pcolor(col_data, edgecolors='k', linewidths=1.5, cmap=plt.cm.get_cmap('tab20c'))
        new_elem = np.array([elem[i] for elem in all_data]).mean(0)
        pos_x, pos_y, dir_x, dir_y = getVec(new_elem)
        ax[i, col].quiver(pos_x, pos_y, dir_x, dir_y, angles='xy', scale_units='xy', scale=1.8, linewidth=0.5)
        plt.setp(ax[i, col].spines.values(), linewidth=1.5, color='k')
        plt.xlim(0, chain_len+2)
        plt.ylim(0, num_chains)
        plt.xticks(range(chain_len+2))
        plt.yticks(range(num_chains))
        plt.grid(which='major', axis='both', linestyle="-", linewidth=1.5, color='black')
        ax[i, col].set_yticklabels([])
        ax[i, col].set_xticklabels([])
plt.show()
```
